```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
with open('data.txt', 'r') as file:
  text = file.read()

text
```

```python
# Text data processing
text = text.lower()
text = nltk.sent_tokenize(text)
text
```

```python
# Tokenization
text2 = 'Here is my second presentation. I feel happier than the first one.'
tokens = nltk.word_tokenize(text2)
tokens
```

```python
sent = nltk.sent_tokenize(text2)
print('Sentence #:',len(sent))
sent
```

```python
# Morphological analysis
nltk.download('wordnet')
tokeniz = nltk.tokenize.ToktokTokenizer()
morph = nltk.stem.WordNetLemmatizer()

token2 = tokeniz.tokenize(text2)
result = [morph.lemmatize(tok) for tok in token2]
print(result)
```

```python
print(morph.lemmatize('happier','a'))
```

```python
# Part-of-speech tagging

tagging = nltk.pos_tag(tokens)
tagging
```

```python
from nltk.corpus import wordnet
synon = []

for i in wordnet.synsets("language"):
  for j in i.lemmas():
    synon.append(j.name())

print(synon)
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

analyze = SentimentIntensityAnalyzer()
senti = analyze.polarity_scores(text)
print(text)
print(senti)

text3 = 'I like this candy it is so sweet and I feel like in the heaven'
senti2 = analyze.polarity_scores(text3)
print('\n', text3)
print(senti2)
```
